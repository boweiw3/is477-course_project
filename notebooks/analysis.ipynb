```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/merged.csv")
df.head()
```

```python
# Basic Overview
print(df.shape)
print(df.info())
df.describe()
```

```python
# Missing Valus Check
df.isnull().sum().sort_values(ascending=False)

# Duplicate Check
df.duplicated().sum()
```

```python
# Crime Trends Over Time
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x="year", y="violent_crime_rate")
sns.lineplot(data=df, x="year", y="property_crime_rate")
plt.legend(["Violent Crime Rate", "Property Crime Rate"])
plt.title("Crime Rates Over Time")
plt.show()
```

```python
# Education Trends Over Time
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x="year", y="edu_staff_total")
plt.title("Education Staff Total Over Time")
plt.xlabel("Year")
plt.ylabel("Education Staff Total")
plt.show()
```

```python
# Correlation Matrix
corr_cols = [
    'violent_crime_rate',
    'property_crime_rate',
    'homicide_rate',
    'robbery_rate',
    'aggravated_assault_rate',
    'edu_staff_total'
]

plt.figure(figsize=(10,8))
sns.heatmap(df[corr_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix: Crime & Education Variables")
plt.show()
```
